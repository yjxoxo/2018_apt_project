### 여기서부터는 그냥 쭈욱 실행

```python
train_size=int(len(apt_lstm)*0.7)
test_size=len(apt_lstm)-train_size
train, test = apt_lstm[0:train_size, :], apt_lstm[train_size:len(apt_lstm),:]
```

```python
len(train)
```

```python
len(test)
```

```python
def seq2dataset(apt, window_size):
    dataset_X = []
    dataset_Y = []
    
    for i in range(len(apt)-window_size):
        subset = apt[i:(i+window_size+1)]
        
        for si in range(len(subset)-1):
            tmp=subset[si]           
            dataset_X.append(tmp)

        dataset_Y.append(subset[window_size][0])
        
    return np.array(dataset_X), np.array(dataset_Y)
```

```python
x_train, y_train = seq2dataset(train, window_size = 3)
x_test, y_test = seq2dataset(test, window_size = 3)
```

```python
x_train = np.reshape(x_train, (72, 3., 8))
x_test = np.reshape(x_test, (30, 3, 8))
```

```python
model = Sequential()
model.add(LSTM(200, batch_input_shape=(1, 3, 8), stateful=True, return_sequences=True))
model.add(LSTM(150, stateful=True, return_sequences=True))
model.add(LSTM(50, stateful=True, return_sequences=False))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
```

```python
num_epochs =700
```

```python
history = LossHistory() # 손실 이력 객체 생성
history.init()
```

```python
import os
from keras.callbacks import ModelCheckpoint
from keras.callbacks import EarlyStopping

model_save_folder_path = './model'
if not os.path.exists(model_save_folder_path):
    os.mkdir(model_save_folder_path)

modelpath= model_save_folder_path + '{epoch:02d}-{loss:4f}.hdf5'
checkpointer = ModelCheckpoint(filepath=modelpath,monitor='loss',verbose=1,save_best_only=True)
early_stopping_callback=EarlyStopping(monitor='loss',patience=50)
```

```python
for epoch_idx in range(num_epochs):
    print ('epochs : ' + str(epoch_idx) )
    model.fit(x_train, y_train, epochs=1, batch_size=1, verbose=2, shuffle=False, callbacks=[history,checkpointer]) # 50 is X.shape[0]
    model.reset_states()
```
